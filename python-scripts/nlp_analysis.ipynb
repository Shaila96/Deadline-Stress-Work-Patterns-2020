```python
mpl.rcParams['font.size']=12                
# mpl.rcParams['savefig.dpi']=100             
mpl.rcParams['figure.subplot.bottom']=.1 


# stop_words = set(STOPWORDS)
stop_words = stopwords.words('english')
stop_words = stop_words + ['need', 'needs', 'NaN']
# stop_words
```

```python
def plot_word_cloud(df, col_name):
    wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stop_words,
                          max_words=40,
                          max_font_size=40, 
                          random_state=42
                         ).generate(str(df[col_name]))

#     print(wordcloud)
    plt.figure(1, figsize=(13, 13))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
```

```python
plot_word_cloud(start_day_df, 'what_task_left_for_the_dea')
```

```python
def split_before_and_after_deadline_data(df, col_name):
    before_df = df[df[col_name].str.contains("-")]
    after_df = df[df[col_name].str.contains("\+")]

    return before_df, after_df

end_day_before_deadline_df, end_day_after_deadline_df = split_before_and_after_deadline_data(end_day_df, 'day')
```

```python
def get_sentiment(_str):
        analysis = TextBlob(_str)

        if analysis.sentiment.polarity > 0:
            return 'positive'
        elif analysis.sentiment.polarity == 0:
            return 'neutral'
        else:
            return 'negative'
```

```python
print(get_sentiment("good"))
```

```python
end_day_before_deadline_df['sentiment'] = end_day_before_deadline_df.apply(lambda row: get_sentiment(row.please_describe_how_you_fe), axis=1)
end_day_before_deadline_df['polarity'] = end_day_before_deadline_df.apply(lambda row: TextBlob(row.please_describe_how_you_fe).sentiment.polarity, axis=1)
```
